## Dataset Overview

The dataset contains 20000 rows and  27 columns representing Indian Personal Finance and Spending Habits.

Columns include: Income, expenses, and potential savings across various categories. The data aims to provide insights into personal financial management and spending patterns.

Data types 25 numerical columns and 2 categorical columns (Occupation and City_Tier).

## Simple description of each column in your dataset

Income & Demographics-

Income: Monthly income in currency units.
Age: Age of the individual.
Dependents: Number of dependents supported by the individual.
Occupation: Type of employment or job role.
City_Tier: A categorical variable representing the living area tier (e.g., Tier 1, Tier 2).


Monthly Expenses-

Categories like Rent, Loan_Repayment, Insurance, Groceries, Transport, Eating_Out, Entertainment, Utilities, Healthcare, Education, and Miscellaneous record various monthly expenses.


Financial Goals & Savings:

Desired_Savings_Percentage and Desired_Savings: Targets for monthly savings.
Disposable_Income: Income remaining after all expenses are accounted for.


Potential Savings:

Includes estimates of potential savings across different spending areas such as Groceries, Transport, Eating_Out, Entertainment, Utilities, Healthcare, Education, and Miscellaneous.

## Checking for outliers

```python
print("Number of outliers as per columns are: \n")
for col in num_cols:
    q1,q3=df[col].quantile([0.25,0.75])
    iqr=q3-q1
    outliers=((df[col]<(q1-1.5*iqr)) | (df[col]>(q3+1.5*iqr))).sum()
    print(f"{col} : {outliers}")
```

```python
### Grouping the columns

# Total Expenses
expense_cols=['Rent', 'Loan_Repayment', 'Insurance', 'Groceries', 'Transport', 'Eating_Out', 
                'Entertainment', 'Utilities', 'Healthcare', 'Education', 'Miscellaneous']
df['Total_Expenses']=df[expense_cols].sum(axis=1)

# Actual Savings and Savings Rate (Income - Expenses) / Income

df['Actual_Savings']=df['Income']-df['Total_Expenses']
df['Actual_Savings_Rate']=(df['Actual_Savings']/df['Income'])*100

# Saving gaps

df['Savings_Gap']=df['Desired_Savings']- df['Actual_Savings']
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.scatterplot(
    x='Income',
    y='Total_Expenses',
    data=df,
    alpha=0.5
)

# Reference line: Income = Expenses
max_val = max(df['Income'].max(), df['Total_Expenses'].max())
plt.plot([0, max_val], [0, max_val], color='red', linestyle='--', label='Income = Expenses')

plt.title('Income vs Total Expenses')
plt.xlabel('Income')
plt.ylabel('Total Expenses')
plt.legend()
plt.tight_layout()
plt.show()

```

```python
# A clear positive relationship is observed between income and total expenses, indicating that spending generally increases as income rises.
# Individuals plotted below the reference line are non-overspenders, as their expenses remain within their income.
#Individuals on or above the reference line are overspenders, indicating higher financial risk and limited or no savings.
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.scatterplot(
    x='Desired_Savings',
    y='Actual_Savings',
    data=df,
    alpha=0.6
)

# Reference line: Desired = Actual
max_val = max(df['Desired_Savings'].max(), df['Actual_Savings'].max())
plt.plot([0, max_val], [0, max_val],
         linestyle='--', color='red', label='Desired = Actual')

plt.title('Desired Savings vs Actual Savings')
plt.xlabel('Desired Savings')
plt.ylabel('Actual Savings')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Shows a strong positive relationship, indicates people intend to save more as their financial capacity grows.
# most of the points lie above the reference line hence most users save less than their desired amount.
# People with ambitious savings goals struggle more to meet them.
# Only a small number of users exceed their savings goals.
```

# Multivariate Analysis

```python
# Defining expense columns
expense_cols = [
    'Rent', 'Loan_Repayment', 'Insurance', 'Groceries',
    'Transport', 'Eating_Out', 'Entertainment',
    'Utilities', 'Healthcare', 'Education', 'Miscellaneous'
]

# Calculating average expense per category
avg_exp = df[expense_cols].mean()

# Selecting top 5 expense categories
top5_cats = avg_exp.sort_values(ascending=False).head(5).index

# Calculating average expenses by city tier for top 5 categories
city_exp = df.groupby('City_Tier', observed=True)[top5_cats].mean()

plt.figure(figsize=(12, 6))
city_exp.plot(kind='bar', cmap='viridis', width=0.8)

plt.title('Top 5 Expense Categories by City Tier', fontsize=16, fontweight='bold', pad=20)
plt.ylabel('Amount (₹)')
plt.xlabel('City Tier')
plt.legend(title='Category', bbox_to_anchor=(1.02, 1), loc='upper left')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
# Rent accounts for the largest share of spending across all city tiers, emphasizing its critical impact on monthly financial planning.
# Cost of living increases with city tier (Tier 1 cities exhibit significantly higher rent and overall expenses, reflecting urban premium costs.)
# Essential expenses are location-independent
# Tier 3 cities offer higher savings potential
```

```python
corr = df[['Income', 'Total_Expenses', 'Actual_Savings',
           'Desired_Savings', 'Savings_Gap']].corr()

sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap of Financial Variables')
plt.tight_layout()
plt.show()
```

```python
# Income is strongly positively correlated with total expenses, indicating lifestyle inflation as earnings increase.
# Income shows a moderate positive correlation with actual savings, suggesting savings grow with income but not proportionally.
# Total expenses have a negative correlation with actual savings.
# A clear mismatch exists between desired and actual savings.
```

```python
# As people earn more, they tend to spend more, so their savings do not grow as much as they expect.
```
